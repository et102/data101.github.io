### Examining quality of two samples  
#### Sampling criteria:  
1. Sample 1:  
 The movie must have had at least 30 fan ratings on Fandango's website at the time of sampling (Aug. 24, 2015).  
 The movie must have had tickets on sale in 2015.
2. Sample 2:  
The movie must have been released in 2016 or later.  
The movie must have had a considerable number of votes and reviews (unclear how many from the README.md or from the data).

However, it is possible to only see if any **"differences"** in two samples.

#### Clean criteria:  
As we only need to find out the differences, below is cleaning criteria for datasets:

### Comparing distribution shape

#### FiveThirtyEight (FTE) visualization for distribution plot

```python
plt.style.use('fivethirtyeight')


before_revised_2015['Fandango_Stars'].plot.kde(label = '2015', legend = True, figsize = (8,5.5))

after_revised_2016['fandango'].plot.kde(label = '2016', legend = True,)

plt.title("Distribution comparison for Fandango ratings\n(2015 vs 2016)",
          y = 1.05) #  y parameter pads the title upward
plt.xlabel('Stars')

plt.xlim(0,5) # because ratings start at 0 and end at 5
plt.xticks(arange(0,5.1,.5))
plt.show()
```

#### Findings:  
1. Both is strongly left skewed.
2. Distribution of 2016 is slightly left shifted, showing the rating is down after the article

#### Frequency distribution tables 

```python
print(before_revised_2015['Fandango_Stars'].value_counts(normalize = True).sort_index() * 100)
print(after_revised_2016['fandango'].value_counts(normalize = True).sort_index() * 100)
```

#### Findings: 
1. High ratings(4.5 & 5) are below 1% in 2016 
2. Minimum rating (2.5) appears in 2016, which is 0 in 2015

### Statistical analysis

```python
mean_2015 = before_revised_2015['Fandango_Stars'].mean()
mean_2016 = after_revised_2016['fandango'].mean()

median_2015 = before_revised_2015['Fandango_Stars'].median()
median_2016 = after_revised_2016['fandango'].median()

mode_2015 = before_revised_2015['Fandango_Stars'].mode()[0]
mode_2016 = after_revised_2016['fandango'].mode()[0]

## creating panda data frame
sta_summary = pd.DataFrame()
sta_summary['2015'] = [mean_2015, median_2015, mode_2015]
sta_summary['2016'] = [mean_2016, median_2016, mode_2016]
sta_summary.index = ['mean', 'median', 'mode']
sta_summary
```

#### FiveThirtyEight (FTE) visualization for distribution statistical summary

```python
plt.style.use('fivethirtyeight')

sta_summary['2015'].plot.bar(color = '#0066AF', align = 'center', label = '2015', width = .25,rot = 0, figsize = (8,5))
sta_summary['2016'].plot.bar(color = '#CC15B0', align = 'edge', label = '2016', width = .25)

plt.title('Summary statistics: 2015 vs 2016', y = 1.05)
plt.ylim(0,5.5)
plt.yticks(arange(0,5.5,.5))
plt.ylabel('Stars')
plt.legend(framealpha = 1, loc = 'upper center')
plt.show()
```

### Conclusion
The difference between 2015 and 2016 is confirmed but we are not sure what changes have been made by Fandango to lower the overall rating.
